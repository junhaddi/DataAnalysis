```python
import csv
f = open('연간공고육비.csv', encoding='cp949')
data = csv.reader(f)

koreaElement = []
koreaMiddle = []
koreaHigh = []

for row in data:
    if '한국' == row[0]:
        for i in range(1, len(row)):
            if i < 4:
                koreaElement.append(int(row[i]))
            elif i < 7:
                koreaMiddle.append(int(row[i]))
            else:
                koreaHigh.append(int(row[i]))
        break

import matplotlib.pyplot as plt
plt.rc('font', family = 'Malgun Gothic')
plt.figure(figsize = (3,3), dpi = 150)
plt.plot([2012, 2013, 2014], koreaElement, label = '초등교육')
plt.plot([2012, 2013, 2014], koreaMiddle, label = '중등교육')
plt.plot([2012, 2013, 2014], koreaHigh, label = '고등교육')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05),
          fancybox=True, shadow=True, ncol=5)

plt.title('대한민국 연간 공교육비')
plt.show()

#TODO 막대 그래프 형식으로 2014년 전세계 공교육비 (초등,중등,고등 교육 width 사용해서 구현해보기)
```
